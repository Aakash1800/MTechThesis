```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
gsd = []
figure = []


for i in range(0,25):
    if __name__ == '__main__':
        fig = plt.gcf()
        fig.set_size_inches(12, 8)
        gsd.append(pd.read_csv('imagemeasurements'+str(i)+'.csv'))
        figure.append(sns.histplot(gsd[i].equivalent_diameter,kde = True))
        plt.xlabel('Equivalent Diameter for Grain')
        plt.ylabel('Count')
        plt.title('Grain Size Distribution '+str(i),loc='right',color='red')
        plt.show()
        
        
```

```python
mean = []
std = []
for i in range(0,25):
    x = (gsd[i])['equivalent_diameter'].mean()
    y = (gsd[i])['equivalent_diameter'].std()
    mean.append(x)
    std.append(y)
```

```python
img_shape = [1000,1000]     #taken from earlier program
pixel_to_um = 3
sum_of_area = []
for i in range(0,25):
    z = (gsd[i])['Area'].sum()
    sum_of_area.append(z)
volumefraction = []
for k in range(0,25):
        volumefraction.append(sum_of_area[i]/(img_shape[0]*img_shape[1]*pixel_to_um*pixel_to_um)) 

        
```

```python
gsd_list = []
for i in range(0,25):
    gsd_list.append('gsd'+str(i))
gsd_data = pd.DataFrame({'Mean':mean,'Std':std,'Volume Fraction':volumefraction},index=gsd_list,columns = ['Mean','Std','Volume Fraction'])

gsd_data
```

```python
sum = 0
for i in mean:
    sum = sum + i
Avg_grain_size_for_all = sum/len(mean)
sum_std = 0
for i in std:
    sum_std = sum_std + i
std_of_grain_size_for_all = sum_std/len(std)
vf = 0
for i in volumefraction:
    vf = vf + i
vfof_grain_size_for_all = vf/len(volumefraction)
std_for_mean = 0
for i in mean:
    std_for_mean = std_for_mean + ((i-Avg_grain_size_for_all)*(i-Avg_grain_size_for_all))
std_for_mean = ((std_for_mean/len(mean))**0.5)
gsd_data_for_all = pd.DataFrame({'Avg of Mean for All':Avg_grain_size_for_all,'Std for Avg of Mean':std_for_mean,'Avg of Std for All':std_of_grain_size_for_all,'Avg of Volume fraction':vfof_grain_size_for_all},index=[1],columns = ['Avg of Mean for All','Std for Avg of Mean','Avg of Std for All','Avg of Volume fraction'])  
gsd_data_for_all  
```
